```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
x = np.arange(1, 21, 1)
```

```python
plt.xlabel('k')
plt.ylabel('PMF')
plt.title('Geometric Distribution')
plt.xticks(x)

for p in [0.5, 0.1]:
    rv = stats.geom(p)
    plt.plot(x, rv.pmf(x), marker='o', label=f'p={p}')
plt.legend()
plt.show()
```
